```python
import seaborn as sns
import matplotlib.pyplot as plt

#设置字体为中文字体的路径
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts//Arial Unicode.ttf',size=10)


df = sns.load_dataset('iris')


sns.boxplot(x=df["sepal_length"], y=df["species"] );
_ = plt.xlabel('花瓣长度 (cm)', fontproperties=font, fontsize=16)
_ = plt.ylabel('物种', fontproperties=font, fontsize=16)
plt.show()
```
